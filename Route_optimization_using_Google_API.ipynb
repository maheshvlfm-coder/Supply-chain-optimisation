## Geocode addresses

### Subtask:
For each address in the DataFrame, use a geocoding service to get the latitude and longitude coordinates.


**Reasoning**:
Define a function to geocode addresses and apply it to the 'Address' column of the DataFrame to get latitude and longitude coordinates.



```python
import requests

def geocode_address(address):
    """Geocodes an address using the OpenStreetMap Nominatim API."""
    try:
        url = f"https://nominatim.openstreetmap.org/search?q={requests.utils.quote(address)}&format=json"
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for bad status codes
        data = response.json()
        if data:
            return (float(data[0]['lat']), float(data[0]['lon']))
        else:
            return (None, None)
    except requests.exceptions.RequestException as e:
        print(f"Error geocoding address '{address}': {e}")
        return (None, None)
    except Exception as e:
        print(f"An unexpected error occurred for address '{address}': {e}")
        return (None, None)

df['Coordinates'] = df['Address'].apply(geocode_address)
display(df)
```

### Subtask:
Geocode addresses using the Google Geocoding API.

**Reasoning**:
Modify the geocode_address function to use the Google Geocoding API and apply it to the 'Address' column of the DataFrame to get latitude and longitude coordinates.

```python
import requests
import time
from google.colab import userdata

def geocode_address_google(address):
    """Geocodes an address using the Google Geocoding API."""
    api_key = userdata.get('GOOGLE_API_KEY')
    if not api_key:
        print("Error: Google API key not found in Colab secrets.")
        return (None, None)

    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    params = {
        "address": address,
        "key": api_key
    }

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Raise an exception for bad status codes
        data = response.json()

        if data["status"] == "OK":
            location = data["results"][0]["geometry"]["location"]
            return (location["lat"], location["lng"])
        elif data["status"] == "ZERO_RESULTS":
            print(f"Geocoding failed for address '{address}': No results found.")
            return (None, None)
        else:
            print(f"Geocoding failed for address '{address}': {data['status']}")
            return (None, None)

    except requests.exceptions.RequestException as e:
        print(f"Error geocoding address '{address}': {e}")
        return (None, None)
    except Exception as e:
        print(f"An unexpected error occurred for address '{address}': {e}")
        return (None, None)
    finally:
        # Add a small delay to avoid hitting rate limits
        time.sleep(0.1)

df['Coordinates'] = df['Address'].apply(geocode_address_google)
display(df)
```

## Use a routing service

### Subtask:
Utilize a routing service (like OSRM, Google Directions API, etc.) to get the detailed geometry of the route between consecutive points in the optimal order.


**Reasoning**:
Define a function to get the route geometry using OSRM and apply it to the ordered coordinates.



```python
import requests

def get_route_geometry(coordinates):
    """Gets the route geometry between a list of coordinates using OSRM."""
    base_url = "http://router.project-osrm.org/route/v1/driving/"
    route_geometry = []

    for i in range(len(coordinates) - 1):
        start_lat, start_lon = coordinates[i]
        end_lat, end_lon = coordinates[i+1]

        # OSRM expects coordinates in 'longitude,latitude' format
        osrm_coords = f"{start_lon},{start_lat};{end_lon},{end_lat}"
        url = f"{base_url}{osrm_coords}?overview=full&geometries=geojson"

        try:
            response = requests.get(url)
            response.raise_for_status()  # Raise an exception for bad status codes
            data = response.json()

            if data and data['code'] == 'Ok':
                # Extract the geometry from the response
                segment_geometry = data['routes'][0]['geometry']['coordinates']
                # OSRM returns [lon, lat], convert to [lat, lon]
                segment_geometry = [(coord[1], coord[0]) for coord in segment_geometry]
                route_geometry.extend(segment_geometry)
            else:
                print(f"OSRM routing failed for segment {i} to {i+1}: {data.get('code', 'Unknown error')}")
                return None # Return None for the whole route if any segment fails

        except requests.exceptions.RequestException as e:
            print(f"Error getting route geometry for segment {i} to {i+1}: {e}")
            return None # Return None for the whole route if any segment fails
        except Exception as e:
            print(f"An unexpected error occurred for segment {i} to {i+1}: {e}")
            return None # Return None for the whole route if any segment fails

    return route_geometry

# Get the detailed route geometry for the optimal path
detailed_route_geometry = get_route_geometry(ordered_coordinates)

if detailed_route_geometry:
    print("Detailed Route Geometry (first 10 points):", detailed_route_geometry[:10])
    print("Total points in route geometry:", len(detailed_route_geometry))
else:
    print("Failed to retrieve detailed route geometry.")
```

## Show journey legs

### Subtask:
Create a table showing each leg of the optimal route with start and end locations and the distance of each leg.

**Reasoning**:
Iterate through the optimal route indices, get the corresponding locations, calculate the distance between consecutive locations using the distance matrix, and store this information in a list of dictionaries. Then, create a pandas DataFrame from this list and display it.

```python
import pandas as pd

journey_legs_data = []
for i in range(len(optimal_route) - 1):
    start_index = optimal_route[i]
    end_index = optimal_route[i+1]

    start_location = df.loc[start_index, 'Location']
    end_location = df.loc[end_index, 'Location']

    # Get distance from the calculated distance matrix
    leg_distance = distance_matrix[start_index][end_index]

    journey_legs_data.append({
        'Leg': i + 1,
        'Start Location (Index)': f"{start_location} ({start_index})",
        'End Location (Index)': f"{end_location} ({end_index})",
        'Distance (km)': leg_distance
    })

# Add the final leg from the last location back to the starting location (TSP is a cycle)
start_index = optimal_route[-1]
end_index = optimal_route[0]

start_location = df.loc[start_index, 'Location']
end_location = df.loc[end_index, 'Location']
leg_distance = distance_matrix[start_index][end_index]

journey_legs_data.append({
    'Leg': len(optimal_route),
    'Start Location (Index)': f"{start_location} ({start_index})",
    'End Location (Index)': f"{end_location} ({end_index})",
    'Distance (km)': leg_distance
})


journey_legs_df = pd.DataFrame(journey_legs_data)
display(journey_legs_df)
```
